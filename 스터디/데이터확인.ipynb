### RFM 구하기

```python
most_recent_dates = purchase_df.groupby('user_id')['event_time'].max()
most_recent_dates
```

```python
unique_user_session_cnt = purchase_df.groupby('user_id')['user_session'].nunique()
unique_user_session_cnt
```

```python
# 구매 기록 확인 -10달러

dff[dff['user_id']==621854037]
```

```python
most_recent_dates_df = most_recent_dates.reset_index(name ='Recency')
unique_user_session_cnt_df = unique_user_session_cnt.reset_index(name='Frequency')
total_revenue_by_user_df = total_revenue_by_user.reset_index(name='Monetary')

RFM = pd.merge(most_recent_dates_df, unique_user_session_cnt_df, on='user_id', how='outer')
RFM = pd.merge(RFM, total_revenue_by_user_df, on='user_id', how='outer')

RFM
```

```python
RFM.info()
```

```python
today_date = pd.to_datetime('2020-03-01').tz_localize('UTC')

rfm = RFM.groupby('user_id').agg({'Recency': lambda x: (today_date - x.max()).days}).reset_index()
RFM = pd.merge(RFM, rfm, on='user_id', how='left')
RFM
```
