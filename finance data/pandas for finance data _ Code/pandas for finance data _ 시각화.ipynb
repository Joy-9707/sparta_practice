## 두개 이상의 그래프

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df_samsung = pd.read_excel('/Users/hj/Documents/무제/finance data/Data/samsung_naver_excel.xlsx', sheet_name=0, index_col = 'Date')
df_samsung
```

```python
fig, axs = plt.subplots(2,1, figsize = (15,10))

axs[0].plot(df_samsung['Close']['2018-05-04':], color = 'blue')

axs[0].grid(True)

axs[0].set(xlabel = "Date", ylabel = 'Price', title = 'Time Series of Samsung Price')

axs[0].plot(df_samsung['Open']['2018-05-04':], color = 'red')

axs[1].scatter(df_samsung.index, df_samsung['Volume'], color = 'red')

axs[1].grid(True)

axs[1].set(xlabel = "Date", ylabel = 'Volume')

plt.show()
```

## Option

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.rcParams["figure.figsize"] = (20,15)
plt.rc('font', family = 'AppleGothic')
plt.rcParams["font.size"] = 15
plt.rc('axes', unicode_minus = False)


df = pd.read_excel('/Users/hj/Documents/무제/finance data/Data/samsung_kospi_excel.xlsx', sheet_name=0, index_col = 'Date')
df = df.loc['2018-01-02':]
df
```

```python
df.index[df['Close'].argmax()]
```

```python
fig, axes = plt.subplots(3,1, sharex = False)  # sharex = True 하면 x축 하나만 나타남 

axes[0].plot(df['Close'], color = 'red', linewidth = 2)
axes[0].plot(df['Close'].rolling(window = 30).mean(), color = 'blue', linewidth = 3)  # 선그래프 굵기 linewidth
axes[0].axhline(df['Close'].mean(), linestyle = '--', linewidth = 5)  # 종가 평균 기준선 만들기 
axes[0].scatter(df.index[df['Close'].argmax()], df['Close'].max(), marker = 'o',s = 500)  # 전고점 표시 
axes[0].legend(['Close', 'Rolling Mean 30days'])
axes[0].set_title("Samsung Stock")

axes[1].plot(df['Change']*100, color ='blue', linewidth = 0.8)
axes[1].axhline(0, linestyle = '--', linewidth = 4, color = 'red')
axes[1].tick_params(axis = 'x', labelrotation = 45)
axes[1].set_title("수익률")

axes[2].fill_between(df.index, df['Volume'], color = 'green', alpha = 1)
axes[2].set_title("거래량")
axes[2].axvline(df.index[df['Volume'].argmax()], color = 'red', linewidth = 5, linestyle = '--')  # 세로선

plt.tight_layout()
plt.show()
```
