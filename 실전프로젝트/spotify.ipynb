```python
# 20대 이탈률 (프리미엄 사용중이나, 앞으로 구독의향 없는 집단)

twenty_bounce = twenty[(twenty['spotify_subscription_plan']=='Premium (paid subscription)') & (twenty['premium_sub_willingness']=='No')]
twenty_bounce

## 사용중인 프리미엄 구독 종류가, 듀오, 패밀리 플랜
```

```python
# 20대 이탈예상 고객의 음악 추천 평가 점수 
twenty_bounce['music_recc_rating'].mean()

# 평균 3.5보다 낮은 수치
```

```python
twenty_bounce.groupby('preferred_listening_content').size()
```

```python
twenty_bounce.groupby('fav_music_genre').size()
```

```python
twenty_bounce.groupby('pod_variety_satisfaction').size()
```
