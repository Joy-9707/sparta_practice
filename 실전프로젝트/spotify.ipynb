```python
df['premium_price'] = df['preffered_premium_plan'].str.split('/').str[0].str.extract('(\d+)')
df.loc[:,['preffered_premium_plan', 'premium_price']].head()
```

```python
# frequency 구하기 
# period 구간 별 임의 숫자 지정 

def random_numbers(df):
    df['Frequency'] = np.nan
    
    period_mapping = {
        'Less than 6 months' : lambda: np.random.randint(1,6),
        '6 months to 1 year' : lambda: np.random.randint(6,12),
        '1 year to 2 years' : lambda: np.random.randint(12,24),
        'More than 2 years' : lambda: np.random.randint(24,36)
    }
    
    for index, row in df.iterrows():
        period = row['spotify_usage_period']
        if period in period_mapping:
            df.at[index, 'Frequency'] = period_mapping[period]()
        
    return df
```

```python
df = random_numbers(df)
df.loc[:,['spotify_usage_period','Frequency']].head()
```

```python
# monetary 계산 

def calculate_price(df):
    df['purchased'] = None
    
    for index, row in df.iterrows():
        if (row['spotify_subscription_plan'] == 'Premium (paid subscription)' and row['premium_sub_willingness'] == 'Yes'):
            df.at[index, 'purchased'] = row['premium_price']
        elif(row['spotify_subscription_plan'] == 'Premium (paid subscription)' and row['premium_sub_willingness'] == 'No'):
            df.at[index, 'purchased'] = row['premium_price']  # (17/24)
        elif(row['spotify_subscription_plan'] == 'Free (ad-supported)'):
            df.at[index, 'purchased'] = 0
            
    df['purchased'].fillna(value=119, inplace=True)
    # mask = (df['spotify_subscription_plan'] == 'Free (ad-supported)' & (df['premium_sub_willingness'] == 'Yes'))
    
    # df.loc[mask, 'purchased'] = df.loc[mask, 'premium_price']
    
    return df
```

```python
df = calculate_price(df)
df.loc[:,['spotify_subscription_plan','premium_sub_willingness', 'preffered_premium_plan', 'purchased']]
```

```python
def calculate_monetary(df):
    
    df['purchased'] = pd.to_numeric(df['purchased'], errors='coerce')
    df['Frequency'] = pd.to_numeric(df['Frequency'], errors='coerce')
    
    df['Monetary'] = df['purchased'] * df['Frequency']
    
    return df

df = calculate_monetary(df)
```

```python
df.loc[:,['Age', 'Gender', 'spotify_usage_period', 'spotify_subscription_plan', 'premium_sub_willingness','preffered_premium_plan','Frequency','purchased','Monetary']].head(50)
```

```python
def calculate_willing_subscribe(df):
    
    df['willing_premium'] = np.where(df['premium_sub_willingness'] == 'Yes', 1, 0)
    
    return df

df = calculate_willing_subscribe(df)
```

```python
import sklearn
from  sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X = df[['Frequency','Monetary']]
y = df['willing_premium']
y
```

```python
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X,y)

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)

# 예측도 확인
```

```python
coef = pd.DataFrame({'features' : X.columns, 'coefficient' : model.coef_[0]})
coef

# 상관관계가 거의 없음...? 그나마 높은것 Frequency
```

```python
coef.iloc[1,1]
```

- 백분위수

```python
a1, a2, a3 = np.percentile(df['Frequency'], [30,50,70])
a1, a2, a3
```

```python
def percent(x):
    if x <= a1:
        return 1
    elif x > a1 and x <= a2:
        return 2
    elif x > a2 and x <= a3:
        return 3
    elif x > a3:
        return 4
```

```python
df['Frequency_score'] = df['Frequency'].apply(percent)*-coef.iloc[0,1]
# *-coef.iloc[0,1]
df
```

```python
df['Frequency_score'].value_counts()
```

```python
b1, b2, b3 = np.percentile(df['Monetary'], [80,83,92])
b1, b2, b3
```

```python
def percent(x):
    if x <= b1:
        return 1
    elif x > b1 and x <= b2:
        return 2
    elif x > b2 and x <= b3:
        return 3
    elif x > b3:
        return 4
```

```python
df['Monetary_score'] = df['Monetary'].apply(percent)*coef.iloc[1,1]
df
```

```python
df['total_score'] = df['Frequency_score'] + df['Monetary_score']
df
```

- total_score 기반으로 5그룹으로 분류

```python
t1, t2, t3, t4 = np.percentile(df['total_score'], [40,50,65,90])
t1, t2, t3, t4
```

```python
def level(x):
    if x <= t1:
        return 5
    elif x > t1 and x <= t2:
        return 4
    elif x > t2 and x <= t3:
        return 3
    elif x > t3 and x <= t4:
        return 2
    elif x > t4:
        return 1
```

```python
df['level'] = df['total_score'].apply(level)
df
```
